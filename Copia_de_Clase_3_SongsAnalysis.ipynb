```python
promedio_popularidad = data['popularity'].mean()
print("Promedio de popularidad:", promedio_popularidad)
```

2.Cual es el artista más popular

```python
artista_mas_popular = data.groupby('artist')['popularity'].mean().idxmax()
print("Artista más popular:", artista_mas_popular)
```

3. Cual es el genero con mayor popularidad y cual es el genero mas popular

```python
genero_mas_popular = data.groupby('genre')['popularity'].mean().idxmax()
print("Género más popular:", genero_mas_popular)
```

```python
data.groupby('genre')['popularity'].mean().sort_values(ascending=False).head(1)
```

4. Cual es el artista que tiene mas canciones con popularidades arriba de 70

```python
data[data["popularity"]>70].groupby("artist").count().sort_values("popularity",ascending=False).head(1)
```

5. La cancion mas popular

```python
data[data["popularity"]==data["popularity"].max()]
```
